```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
iris = load_iris()

import pandas as pd
X = pd.DataFrame(iris.data)
X.columns = ['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width']
y = pd.DataFrame(iris.target)
y.columns = ['Targets']

import numpy as np
colormap = np.array(['red', 'lime', 'black'])

from sklearn.cluster import KMeans
model = KMeans(n_clusters=3)
model.fit(X)
model.labels_
plt.subplot(1, 2, 1)
plt.scatter(X.Petal_Length, X.Petal_Width, c=colormap[y.Targets], s=40)
plt.title('Real Classification')

plt.subplot(1, 2, 2)
plt.scatter(X.Petal_Length, X.Petal_Width, c=colormap[model.labels_], s=40)
plt.title('K Mean Classification')

import sklearn.metrics as sm
print(sm.accuracy_score(y, model.labels_))

from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components=3)
gmm.fit(iris.data)
y_cluster_gmm = gmm.predict(iris.data)
y_cluster_gmm
plt.subplot(1, 2, 1)
plt.scatter(X.Petal_Length, X.Petal_Width, c=colormap[y_cluster_gmm], s=40)
plt.title('GMM Classification')

print(sm.accuracy_score(y, y_cluster_gmm))
print(sm.confusion_matrix(y, y_cluster_gmm))
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
import pandas as pd
X = pd.DataFrame(iris.data)
X.columns = ['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width']
y = pd.DataFrame(iris.target)
y.columns = ['Targets']
```

```python
import numpy as np
colormap = np.array(['red', 'lime', 'black'])
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3)
model.fit(X)
```

```python
plt.subplot(1, 2, 1)
plt.scatter(X.Petal_Length, X.Petal_Width, c=colormap[y.Targets], s=40)
plt.title('Real Classification')
plt.subplot(1, 2, 2)
plt.scatter(X.Petal_Length, X.Petal_Width, c=colormap[model.labels_], s=40
)
plt.title('K Mean Classification')
predY = np.choose(model.labels_, [0, 1, 2]).astype(np.int64)
print (predY)
```

```python
plt.figure(figsize=(14,7))
# Create a colormap
colormap = np.array(['red', 'lime', 'black'])
# Plot Orginal
plt.subplot(1, 2, 1)
plt.scatter(X.Petal_Length, X.Petal_Width, c=colormap[y.Targets], s=40)
plt.title('Real Classification')
# Plot Predicted with corrected values
plt.subplot(1, 2, 2)
plt.scatter(X.Petal_Length,X.Petal_Width, c=colormap[predY], s=40)
plt.title('K Mean Classification')
```

```python
import sklearn.metrics as sm
print(sm.accuracy_score(y, model.labels_))
print(sm.confusion_matrix(y, model.labels_))

```

```python
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components=3)
gmm.fit(X)
```

```python
y_cluster_gmm = gmm.predict(iris.data)
y_cluster_gmm
```

```python
sm.accuracy_score(y, y_cluster_gmm)
```

```python
sm.confusion_matrix(y, y_cluster_gmm)
```
